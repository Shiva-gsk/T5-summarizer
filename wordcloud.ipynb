```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./output.csv')
```

```python
from wordcloud import WordCloud

def plot_wordcloud(data, text_column):

        text = " ".join(data[text_column].astype(str)) # Convert to string
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

        # Display the word cloud
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title("Word Cloud for All Feedback")
        plt.show()

# Example Usage
plot_wordcloud(df, text_column='Summary')
```

 We can see the most occuring words in our db


```python
plot_wordcloud(df, text_column='Text')
```
